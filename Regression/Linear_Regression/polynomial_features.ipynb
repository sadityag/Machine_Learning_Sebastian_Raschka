Linear regression can be extended to Polynomial features.

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([258.0, 270.0, 294.0, 320.0, 342.0, 368.0, 396.0, 446.0, 480.0, 586.0])[:, np.newaxis]
y = np.array([236.4, 234.4, 252.8, 298.6, 314.2, 342.2, 360.8, 368.0, 391.2, 390.8])
lr = LinearRegression()
pr = LinearRegression()
quadratic = PolynomialFeatures(degree=2)
X_quad = quadratic.fit_transform(X)
```

```python
X_points = np.arange(250, 600, 10)[:, np.newaxis]

lr.fit(X, y)
pr.fit(X_quad, y)

y_linear_fit = lr.predict(X_points)
y_poly_fit = pr.predict(quadratic.transform(X_points))
```

```python
plt.scatter(X, y, label='Training points', c='steelblue')
plt.plot(X_points, y_linear_fit, label='Linear fit', linestyle='--', c='red')
plt.plot(X_points, y_poly_fit, label='Quadratic fit', c='limegreen')
plt.xlabel('Explanatory variable')
plt.ylabel('Predicted or known target values')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print(f"Linear fit \n MSE: {mean_squared_error(y_true=y, y_pred=lr.predict(X)):.3f} \n R^2: {r2_score(y_true=y, y_pred=lr.predict(X)):.3f}")
print(f"Quadratic fit \n MSE: {mean_squared_error(y_true=y, y_pred=pr.predict(X_quad)):.3f} \n R^2: {r2_score(y_true=y, y_pred=pr.predict(X_quad)):.3f}")
```
